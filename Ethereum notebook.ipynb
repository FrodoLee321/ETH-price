```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ETH-USD.csv')
print (data.shape)
data.info()
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Year'] = data['Date'].dt.strftime('%Y')
data['Month'] = data['Date'].dt.strftime('%m')
data['Day'] = data['Date'].dt.strftime('%d')
```

```python
a = ['Open', 'Close', 'Low', 'High', 'Year', 'Month']
c = data[a].groupby(['Year', 'Month']).sum()
c
```

```python
c.plot(figsize=(20, 10), grid=True, marker='*')
plt.show()
```

```python
c.plot(kind='bar', figsize=(18, 9))
plt.show()
```

```python
c.plot(kind='area', figsize=(18, 9))
plt.show()
```
